```python
 class Patient:
    def __init__(self, name, age, gender, disease):
        self.name = name
        self.age = age
        self.gender = gender
        self.disease = disease

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Gender: {self.gender}, Disease: {self.disease}"


class Doctor:
    def __init__(self, name, speciality):
        self.name = name
        self.speciality = speciality

    def __str__(self):
        return f"Name: {self.name}, Speciality: {self.speciality}"


class Hospital:
    def __init__(self):
        self.patients = []
        self.doctors = []

    def add_patient(self, patient):
        self.patients.append(patient)

    def add_doctor(self, doctor):
        self.doctors.append(doctor)

    def list_patients(self):
        if not self.patients:
            print("No patients found.")
            return
        for patient in self.patients:
            print(patient)

    def list_doctors(self):
        if not self.doctors:
            print("No doctors found.")
            return
        for doctor in self.doctors:
            print(doctor)

    def search_patient(self, name):
        for patient in self.patients:
            if patient.name.lower() == name.lower():
                return patient
        return None


def main():
    hospital = Hospital()

    while True:
        print("\n--- Hospital Management System ---")
        print("1. Add Patient")
        print("2. Add Doctor")
        print("3. List Patients")
        print("4. List Doctors")
        print("5. Search Patient")
        print("6. Exit")

        choice = input("Select an option (1-6): ")

        if choice == '1':
            name = input("Enter patient's name: ")
            age = int(input("Enter patient's age: "))
            gender = input("Enter patient's gender: ")
            disease = input("Enter patient's disease: ")
            new_patient = Patient(name, age, gender, disease)
            hospital.add_patient(new_patient)
            print(f"Patient {name} added successfully.")

        elif choice == '2':
            name = input("Enter doctor's name: ")
            speciality = input("Enter doctor's speciality: ")
            new_doctor = Doctor(name, speciality)
            hospital.add_doctor(new_doctor)
            print(f"Doctor {name} added successfully.")

        elif choice == '3':
            print("\nPatients:")
            hospital.list_patients()

        elif choice == '4':
            print("\nDoctors:")
            hospital.list_doctors()

        elif choice == '5':
            patient_to_find = input("Enter the patient's name to search: ")
            found_patient = hospital.search_patient(patient_to_find)
            if found_patient:
                print(f"\nFound patient: {found_patient}")
            else:
                print(f"\nPatient {patient_to_find} not found.")

        elif choice == '6':
            print("Exiting the system.")
            break

        else:
            print("Invalid option. Please try again.")


if __name__ == "__main__":
    main()
```
